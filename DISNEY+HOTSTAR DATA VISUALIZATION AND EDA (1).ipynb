Question 11: Which cast member holds the record for appearing in the most movies within the Hotstar dataset, and what are the genres associated with their appearances? Could you provide insights into this cast member's broad range of roles across different genres?

```python
# Extracting and process cast information
cast_column = 'cast'
exploded_cast = df[cast_column].str.split(', ').explode()

# Creating a DataFrame with cast and genre information
genre_column = 'listed_in'
cast_genre_counts = df[[cast_column, genre_column]].explode(cast_column)

# Group and count cast appearances with genres
cast_genre_counts = cast_genre_counts.groupby([cast_column, genre_column]).size().reset_index(name='appearance_count')

# Identify the most frequent cast member
most_frequent_cast = cast_genre_counts[cast_genre_counts['appearance_count'] == cast_genre_counts['appearance_count'].max()]

# Printing the results
print("Cast member who appeared in the most movies and their associated genres:")
print(most_frequent_cast)
```

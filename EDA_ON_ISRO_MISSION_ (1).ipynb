INFORMATION ABOUT DATASET

The dataset I have consists of information related to ISRO (Indian Space Research Organisation) space missions. Here's a brief overview of the dataset columns:

Serial Number: A unique identifier for each mission.
Name: The name of the mission.
Launch Date: The date on which the rocket was launched for the mission.
Launch Vehicle: The type of rocket used to carry the spacecraft to space.
Orbit Type: The type of orbit in which the spacecraft operates (e.g., Geostationary orbit, Low Earth orbit, etc.).
Application: The specific application or purpose of the mission.
Launcher Type: The type of launcher used for the mission.
Payload: Information about the scientific or technological instrument carried on board the satellite for its specific purpose.
Remarks: Remarks or notes related to the mission.
Mission Type: The type of mission, categorized as "Spacecraft" or "Launch."

Question1:How has the number of ISRO space missions changed over the years?

```python
#converting launch Date column in datetime fromat
isro['Launch Date']= pd.to_datetime(isro['Launch Date'])
```

```python
mission_per_year=isro['Launch Date'].value_counts().sort_index()
```

```python
mission_per_year
```

```python
# CreatING  a time series plot
plt.figure(figsize=(10, 6))
plt.plot(mission_per_year.index, mission_per_year.values, marker='o')
plt.title('Number of ISRO Space Missions Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.grid(True)
plt.show()
```

```python
# Find specific years with higher mission counts
threshold = 10
years_with_high_missions = mission_per_year[mission_per_year >= threshold]
print("Years with higher mission counts:")
print(years_with_high_missions)
```

Question:Can you identify any trends or factors that contribute to mission success or failure based on remarks?

```python

# Exploring unique values in the "Remarks" column
unique_remarks = isro['Remarks'].unique()
print("Unique Remarks:")
for remark in unique_remarks:
    print(remark)
```

```python
#Handling missing values of remarks column
isro['Remarks'] = isro['Remarks'].fillna('Unknown')
```

```python
# Handling missing values in the "Remarks" column
isro['Remarks'] = isro['Remarks'].fillna('Unknown')

# Categorize remarks into Success, Failure, and Other
isro['Mission Outcome'] = isro['Remarks'].apply(lambda x: 'Success' if 'Success' in str(x) else ('Failure' if 'Failure' in str(x) else 'Other'))

# Count the number of missions for each outcome category
outcome_counts = isro['Mission Outcome'].value_counts()

# Display the outcome counts
print("Mission Outcome Counts:")
print(outcome_counts)
```

Question:Which application category has the highest success rate? Are there any applications that tend to have more mission failures?

```python
# Defining categories for different applications
application_categories = {
    'Communication': ['Communication Satellite', 'Communication'],
    'Remote Sensing': ['Earth Observation'],
    'Scientific Research': ['Scientific'],
    'Navigation': ['Navigation'],
    'Interplanetary': ['Planetary Science'],
    'Space Station': ['Space Station'],
    'Technology Development': ['Technology'],
    'Miscellaneous': ['Miscellaneous']
}

# Creating a new column for application categories
isro['Application Category'] = isro['Application'].apply(
    lambda x: next((category for category, apps in application_categories.items() if x in apps), 'Other')
)

# Calculating success rates for each application category
application_success_rates = isro.groupby('Application Category')['Mission Outcome'].apply(lambda x: (x == 'Success').sum() / len(x))

# Displaying the success rates for each application category
print("Success Rates by Application Category:")
print(application_success_rates)

# Counting the number of failures for each application category
application_failure_counts = isro[isro['Mission Outcome'] == 'Failure'].groupby('Application Category')['Mission Outcome'].count()

# Displaying the number of failures for each application category
print("\nNumber of Failures by Application Category:")
print(application_failure_counts)
```

Question:Can you find any correlation between public interest/media coverage and mission success or failure?

```python
#Generating synthetic media overage and sentiment data for illustration
np.random.seed(42)
n_mission=len(isro)
media_coverage=np.random.randint(0,100,n_mission)#example of media coverage data
sentiment=np.random.uniform(-1,1,n_mission)# example sentiment data
```

```python
#Adding synthetic data to the mission dataset
isro['Media Coverage']=media_coverage
isro['Sentiment']=sentiment
```

```python
#Performing correlation analysis
corr_media_success=isro['Media Coverage'].corr(isro['Mission Outcome']=='success')
corr_sentiment_success=isro['Sentiment'].corr(isro['Sentiment']=='success')
```

```python
# Displaying the correlation results
print("Correlation between Media Coverage and Mission Success:", corr_media_success)
print("Correlation between Sentiment and Mission Success:", corr_sentiment_success)
```
